```python
import datasets
import random
import pandas as pd
import numpy as np
from itertools import chain
```

```python
dataset = datasets.load_dataset("squad")

dataset
```

```python
eval_dataset = dataset["validation"]

eval_dataset
```

```python
eval_dataset = pd.DataFrame(eval_dataset)

eval_dataset
```

```python
eval_dataset_q = eval_dataset.copy()
eval_dataset_q["question"] = np.random.default_rng(seed=388).permutation(eval_dataset["question"].values)

eval_dataset_q
```

```python
eval_dataset_qs = datasets.Dataset.from_pandas(eval_dataset_q)

eval_dataset_qs
```

```python
vocab = []
with open("../models/model_init/vocab.txt", encoding="utf-8") as f:
    for line in f.readlines():
        vocab.append(line.strip())

len(vocab)
```

```python
eval_dataset["gold_answer"] = eval_dataset["answers"].apply(lambda d: d["text"][0])

eval_dataset
```

```python
passage_answers = eval_dataset.groupby("context")["gold_answer"].apply(list)
```

```python
def randomize_passage(passage, answers):
    passage_split = passage.split(" ")
    answers_split = [answer.split(" ") for answer in answers]
    random.shuffle(answers_split)
    indices = [0]

    for answer in answers_split:
        indices.append(indices[-1] + len(answer))
    
    passage_random = list(chain.from_iterable(answers_split))

    while len(passage_random) < len(passage_split):
        insert_slot = random.choice(range(len(indices)))
        passage_random.insert(indices[insert_slot], random.choice(vocab))
        for i in range(insert_slot + 1, len(indices)):
            indices[i] += 1
    
    return " ".join(passage_random)
```

```python
eval_dataset[eval_dataset["context"] == passage_answers.index[0]]

# Find answers associated with each passage and randomize them one by one probably
```

```python
random.seed(388)
random_passages = []
for passage in eval_dataset["context"]:
    random_passages.append(randomize_passage(passage, passage_answers[passage]))
```

```python
eval_dataset_p = eval_dataset.copy()

eval_dataset_p["context"] = pd.Series(random_passages)
eval_dataset_p = eval_dataset_p.drop("gold_answer", axis=1)
```

```python
eval_dataset_ps = datasets.Dataset.from_pandas(eval_dataset_p)

eval_dataset_ps
```
